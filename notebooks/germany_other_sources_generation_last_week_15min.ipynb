```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Actual_generation_202212250000_202512250000_Quarterhour.csv", sep=";", low_memory=False)
```

```python
cols_to_convert = [
    "Biomass [MWh] Original resolutions",
    "Hydropower [MWh] Original resolutions",
    "Wind offshore [MWh] Original resolutions",
    "Wind onshore [MWh] Original resolutions",
    "Photovoltaics [MWh] Original resolutions",
    "Other renewable [MWh] Original resolutions",
    "Nuclear [MWh] Original resolutions",
    "Lignite [MWh] Original resolutions",
    "Hard coal [MWh] Original resolutions",
    "Fossil gas [MWh] Original resolutions",
    "Hydro pumped storage [MWh] Original resolutions",
    "Other conventional [MWh] Original resolutions"
]

for col in cols_to_convert:
    # Remove any stray spaces and force conversion to float
    if df[col].dtype == object:
        df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns

df[numeric_cols] = df[numeric_cols] * 4
```

```python
clean_df = df[[
    "Biomass [MWh] Original resolutions",
    "Hydropower [MWh] Original resolutions",
    "Nuclear [MWh] Original resolutions",
    "Lignite [MWh] Original resolutions",
    "Hard coal [MWh] Original resolutions",
    "Fossil gas [MWh] Original resolutions",
    "Hydro pumped storage [MWh] Original resolutions",
    "Other conventional [MWh] Original resolutions"
]].rename(columns={
    "Biomass [MWh] Original resolutions": "germany_biomass_gen_last_week",
    "Hydropower [MWh] Original resolutions": "germany_hydro_passive_gen_last_week",
    "Nuclear [MWh] Original resolutions": "germany_nuclear_gen_last_week",
    "Lignite [MWh] Original resolutions": "germany_lignite_gen_last_week",
    "Hard coal [MWh] Original resolutions": "germany_hard_coal_gen_last_week",
    "Fossil gas [MWh] Original resolutions": "germany_fossil_gas_gen_last_week",
    "Hydro pumped storage [MWh] Original resolutions": "germany_hydro_active_gen_last_week",
    "Other conventional [MWh] Original resolutions": "germany_other_gen_last_week"
})
```

```python
clean_df.to_csv("../data/germany_other_sources_generation_last_week_15min.csv", sep=";", index=False)
```
