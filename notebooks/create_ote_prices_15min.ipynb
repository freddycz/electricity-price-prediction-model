```python
import pandas as pd
```

```python
df = pd.read_csv("../data/ote_2022-12-25_2025-12-31_prices.csv", sep=";", low_memory=False)
```

```python
cols_to_convert = [
    "Marginální cena ČR (EUR/MWh)",
]

for col in cols_to_convert:
    # Remove any stray spaces and force conversion to float
    if df[col].dtype == object:
        df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["Den"] = pd.to_datetime(df["Den"])
```

```python
clean_df = df[["Den", "Perioda", "Časový interval", "Marginální cena ČR (EUR/MWh)"]].rename(columns={
    "Den": "date",
    "Perioda": "period",
    "Časový interval": "time_interval",
    "Marginální cena ČR (EUR/MWh)": "price"
})
```

```python
start_times = clean_df['time_interval'].str.split('-').str[0].str.replace(r'[a-zA-Z]', '', regex=True)
start_dt = pd.to_datetime(start_times, format='%H:%M')
clean_df['is_peak'] = start_dt.dt.hour.between(8, 19)
```

```python
clean_df.to_csv("../data/ote_prices_15min.csv", sep=";", index=False)
```
