```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Actual_generation_202212250000_202512250000_Quarterhour.csv", sep=";", low_memory=False)
```

```python
cols_to_convert = [
    "Photovoltaics [MWh] Original resolutions",
    "Wind offshore [MWh] Original resolutions",
    "Wind onshore [MWh] Original resolutions"
]

for col in cols_to_convert:
    # Remove any stray spaces and force conversion to float
    if df[col].dtype == object:
        df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["Wind"] = (
    df["Wind offshore [MWh] Original resolutions"] + 
    df["Wind onshore [MWh] Original resolutions"]
)
```

```python
df["Wind"] = df["Wind"].round(2)
```

```python
clean_df = df[["Photovoltaics [MWh] Original resolutions", "Wind"]].rename(columns={
    "Photovoltaics [MWh] Original resolutions": "germany_solar_gen_last_week",
    "Wind": "germany_wind_gen_last_week"
})
```

```python
clean_df.to_csv("../data/wind_solar_generation_last_week_15min.csv", sep=";", index=False)
```
