```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/Actual_generation_202212250000_202512250000_Quarterhour.csv", sep=";", low_memory=False)
```

```python
df["Start date"] = pd.to_datetime(df["Start date"], format="%b %d, %Y %I:%M %p")
df["End date"] = pd.to_datetime(df["End date"], format="%b %d, %Y %I:%M %p")
```

```python
cols_to_convert = [
    "Photovoltaics [MWh] Original resolutions",
    "Wind offshore [MWh] Original resolutions",
    "Wind onshore [MWh] Original resolutions"
]

for col in cols_to_convert:
    # Remove any stray spaces and force conversion to float
    if df[col].dtype == object:
        df[col] = df[col].astype(str).str.replace(",", "")
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["Wind"] = (
    df["Wind offshore [MWh] Calculated resolutions"] + 
    df["Wind onshore [MWh] Calculated resolutions"]
)
```

```python
solar_col = "Photovoltaics [MWh] Original resolutions"
wind_col = "Wind"
```

```python
df[solar_col] = df[solar_col]
df[wind_col] = df[wind_col]
```

```python
clean_df = df[["Start date", "End date", "Photovoltaics [MWh] Original resolutions", "Wind"]].rename(columns={
    "Start date": "start_date",
    "End date": "end_date",
    "Photovoltaics [MWh] Original resolutions": "solar_production",
    "Wind": "wind_production"
})
```

```python
clean_df.to_csv("../data/wind_solar_production_last_week_15min.csv", sep=";", index=False)
```
