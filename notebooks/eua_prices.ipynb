```python
import pandas as pd
```

```python
df = pd.read_csv("../data/EUA.csv", sep=",")
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'], dayfirst=True)
```

```python
df.set_index('DateTime', inplace=True)
df.sort_index(inplace=True)
```

```python
start_date = df.index.min() 
end_date = df.index.max()   
full_date_range = pd.date_range(start=start_date, end=end_date, freq='D')
```

```python
df_filled = df.reindex(full_date_range).ffill()
```

```python
df_filled.index.name = 'DateTime'
df_filled.reset_index(inplace=True)
```

```python
df_filled.rename(columns={
    'DateTime': 'date',
    'Cena [EUR/t]': 'eua_price'
}, inplace=True)
```

```python
df_filled.to_csv('../data/eua_prices_complete.csv', index=False, sep=";")
```
